### Conditional Statements

```python
'''
define the function greet_admin

function takes a user's name as a parameter

if the user is Guido then it will return 

"Welcome, Guido" if not then the response is empty.
'''
def greet_admin(user):
    if user == "Guido":
        return "Welcome, Guido."
    
print(greet_admin("Guido"))
print(greet_admin("Alex"))
```

```python
'''
define the function greet_admin

function takes a user's name as a parameter

if the user is Guido then it will return 

"Welcome, Guido" if not then we've added 

alternative statements to return for Bethany or Alex

and an else statement to return something other than None.
'''
def greet_admin(user):
    if user == "Guido":
        return "Welcome, Guido."
    elif user == "Bethany":
        return "Welcome, Bethany."
    elif user == "Alex":
        return "Welcome, Alex."
    else:
        return "You are not authorized."

print(greet_admin("Guido"))
print(greet_admin("Alex"))
print(greet_admin("Grae"))
```

### Loops

```python
'''
initialize two empty lists

loop through a list ranging from 0 to 9

if the number divided by 2 has a remainder of 0
then the number is even and should be appended 
to the even numbers list

else append the number squared to the odd numbers list
'''
even_numbers = []
odd_squares = []

for n in range(10):
    if n % 2 == 0:
        even_numbers.append(n)
    else:
        odd_squares.append(n ** 2)

print(even_numbers)
print(odd_squares)
```

```python
'''
initialize n

create an incremental variable

if n is an even number then enter the loop

update n with the value of n divided by 2

add 1 to divisions to indicate how many times 
we've been through the loop 

once loop reaches an odd number then loop will exit

message will be saved and printed using the final value for n
'''
n = 24
divisions = 0
print("To start, n is {}".format(n))
while n % 2 == 0:
    n = n // 2
    divisions += 1
    print("After {} division(s) n is {}".format(divisions, n))

# Now that our loop is done, let's see the result.
message = "After removing all factors of two, the number is {}"
print(message.format(n))
```

### OOP

```python
'''
create the class Quark
initialize the variables associated with the Quark class
take in parameters for color and flavor and save them to self

define methods in similar way to defining standalone functions

define interact method to swap colors between quarks
when interact is called the other quark's color is saved
as the new self.color and the other quark's color becomes 
the initial self.color

when we run the Quark class we want it to visually display 
something. in this case we want a print statement of the 
quarks colors and flavors

we use the built in __repr__ method to print the
color and flavor of the quark

we save red and up to the first quark 
and blue down to the second quark

when we call the interact method the first quark
switches to blue up and second quark to red down
'''
class Quark(object):

    # This method is automatically called whenever we create a new quark.
    # It sets the color and flavor attributes when we create an instance.
    def __init__(self, color, flavor):
        self.color = color
        self.flavor = flavor

    # Every quark has the same baryon number, so we set this outside the
    # init function.
    baryon_number = 1 / 3

    # This method models the way quarks interact with one another by
    # exchanging color.
    def interact(self, other_quark):
        self.color, other_quark.color = other_quark.color, self.color

    # The repr method controls how the object is represented by the
    # print() function and other representations of the object.
    def __repr__(self):
        return "{} {} quark".format(self.color, self.flavor)

# Now that we have the class set up, let's call Quark() to create two
# actual instances of quark objects.
q1 = Quark("red", "up")
q2 = Quark("blue", "down")

# Print each object to see what they look like.
print("q1 is a {}".format(q1))
print("q2 is a {}".format(q2))

# Test our interact() method by having q1 and q2 interact.
q1.interact(q2)

# Print them out again to see how they changed.
print("Now q1 is a {}".format(q1))
print("Now q2 is a {}".format(q2))

# Test how our object deals with the built-in type() function.
print("q1 is a {} object".format(type(q1)))
```
